```python
aui_info = []

with open('/data/Bodenreider_UMLS_DL/UMLS_VERSIONS/2020AA-ACTIVE/META/MRCONSO.RRF','r') as fp:
    
    for line in fp.readlines():
        line = line.split('|')
        cui = line[0]
        aui = line[7]
        string = line[-5]
        
        aui_info.append({'AUI':aui, 'CUI':cui, 'STR':string})
        
cui2sg = {}

with open('/data/Bodenreider_UMLS_DL/UMLS_VERSIONS/2020AA-ACTIVE/META/MRSTY.RRF','r') as fp:
    
    for line in fp.readlines():
        line = line.split('|')
        cui = line[0]
        sg = line[3]
        cui2sg[cui] = sg
        
cui2aui = {}
aui2cui = {}
aui2str = {}
aui2sg = {}

cui_sg = []
cui_aui = []

for tup in tqdm(aui_info):
    aui = tup['AUI']
    
    cui = tup['CUI']
    string = tup['STR']
    sg = cui2sg[cui]

    auis = cui2aui.get(cui, [])
    auis.append(aui)
    cui2aui[cui] = auis

    aui2cui[aui] = cui
    aui2str[aui] = string
    aui2sg[aui] = sg

    cui_sg.append((cui, sg))
    cui_aui.append((cui, aui))
        
semgroups = pd.read_csv('SemGroups.txt',sep='|',header=None)

semtype2sg = {}

for i, row in semgroups.iterrows():
    
    st = row[3]
    sg = row[1]
    
    semtype2sg[st] = sg
```

```python
f = open('/data/Bodenreider_UMLS_DL/Interns/Bernal/uva_training_data.csv','r')

f.readline()
training_set_dict = {}

for line in tqdm(f,total=173000000):
    line = line.strip()
    _,_,_,_,_,_,h,_,t = line.split('\t')
    
    h_edges = training_set_dict.get(h,set())
    t_edges = training_set_dict.get(t, set())
    
    h_edges.add(t)
    t_edges.add(h)
    
    training_set_dict[h] = h_edges
    training_set_dict[t] = t_edges
```

```python
pickle.dump(testing_data_tuples, open('/data/Bodenreider_UMLS_DL/Interns/Bernal/uva_test_edges.p','wb'))
```

```python
in_training = []

f = open('/data/Bodenreider_UMLS_DL/Interns/Bernal/uva_testing_data.csv','r')

f.readline()

for line in tqdm(f,total=173000000):
    line = line.strip()
    _,_,_,_,_,_,h,_,t = line.split('\t')
    
    h_edges = training_set_dict.get(h,set())
    
    if t in h_edges:
        in_training.append(True)
    else:
        in_training.append(False)
```
